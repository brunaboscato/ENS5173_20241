### 1.1 Numpy arrays

```python
import numpy as np 

# Criando um vetor de dados aleatórios
data = np.random.rand(1000,2)

# Verificando o tamanho do vetor
print(data.shape)
print(np.size(data))

# Print de uma amostra dos dados
print(data[0:10]) # pedindo para mostrar apenas as 10 primeiras linhas
print(data[0:10].shape) # formato 
```

Monte Carlo é uma técnica computacional utilizada para resolver problemas através de métodos estatísticos e da simulação de eventos aleatórios.

```python
# Usando Monte-Carlo
dataMax = 1000
dataMin = 10
data = (dataMax - dataMin)*data + dataMin
print(data[0:10])
# estipulou um dataMax e dataMin, multiplicou os dados pela diferença deles e depois somou o dataMin nos dados
```

```python
# Plotando os dados
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3)
ax[0].plot(data) # plotando os dados completos
ax[1].plot(data[:,0]) # plotando os dados da primeira coluna
ax[2].plot(data[:,1]) # plotando os dados da segunda coluna
print(type(data))
```

DataFrame é uma estrutura de dados bidimensional semelhante a uma tabela.

Indexação é o processo de acessar elementos em uma sequência.

```python
# Colocando o datetime como index
df = df.set_index(df['datetime']) # o índice do DataFrame df será definido como a coluna 'datetime'
df
```

```python
# Criando colunas com ano, mês e dia
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
```

```python
tt = np.arange(0,df.shape[0],1) # criando um array "tt" usando a função np.arange(), que gera valores de 0 até o número de linhas no DataFrame df, com um passo de 1.
print(tt.shape)
df['sazonal'] = (np.cos(tt/(360*4))) # usando a função np.cos() para calcular o cosseno de cada valor em tt dividido por (360*4)
df['timeSeries'] = 10*df['sazonal']+df['random'] # calculando a série temporal completa combinando a componente sazonal (multiplicada por 10) com uma série aleatória já existente no DataFrame df como 'random'
#df['sazonal'].plot()
df['timeSeries'].plot(color='red') # plotando a série temporal completa "timeSeries" usando a função plot() do pandas, com a cor vermelha
```

Este método converte o DataFrame em um arquivo CSV e o salva no caminho especificado.

```python
# Criando uma função
def noiseCreation(path, year): # definindo uma função chamada "noiseCreation"
    """Esta função cria um ruido com sazonalidade.
    Inputs:
        path = asdasdas
        year = asdasdasdas

    Criado por Leonardo
    """
    
    # Criando um dataframe
    df = pd.DataFrame()
    
    # Criando vertor temporal
    dtime = np.arange(np.datetime64(str(year)+"-01-01"), np.datetime64(str(year+1)+"-01-01"), np.timedelta64(1, "h"))
    
    # Monstrando um dataframe
    df['datetime'] = dtime
        
    # Datetime como index do dataframe
    df = df.set_index(df['datetime'])
    
    # Criando colunas com ano, mês e dia
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    
    # Crio uma coluna random
    df['random'] = np.random.rand(df.shape[0])
    
    # Adicionar sazonalidade
    tt = np.arange(0,df.shape[0],1)
    print(tt.shape)
    df['sazonal'] = (np.cos(tt/(360*4))) # sazonalidade é adicionada à série temporal através da coluna "sazonal"
    df['timeSeries'] = 10*df['sazonal']+df['random']
    #df['sazonal'].plot()
    df['timeSeries'].plot(color='red') # plotando a série temporal resultante com a cor vermelha
    df.to_csv(path+'/noise_'+str(year)+'.csv') # salvando o DataFrame em um arquivo CSV com o nome "noise_ano.csv" no caminho especificado
    
    return df
```

```python
# definindo uma variável chamada path, que contém o caminho para o diretório onde os arquivos CSV serão salvos
path = 'C:/Users/bruna/OneDrive/Documentos/UFSC/Controle de Poluicao Atmosferica/ENS5173_20241/Dados' 
df2 = noiseCreation(path,2050) # caminho para o diretório onde os arquivos CSV serão salvos (definido pela variável path), e o ano de 2050
```

```python
# Primeira opção de loop
for ii in range(2002,2010): # valor de ii será atualizado em cada iteração, começando com 2002 e terminando com 2009
    print(ii)
    df = noiseCreation(path,ii)
```

```python
anos = [2001,2030,2032,2053,2060] # lista chamada anos

for ano in anos: # vai iterar sobre cada elemento da lista anos usando um loop "for"
    print(ano)
    if ano > 2030:
        df = noiseCreation(path,ano)

# vai passar em cada ano na lista anos 
# imprimirá o ano atual e, se o ano for maior que 2030, chamará a função noiseCreation, passando o caminho (path) e o ano como argumentos
# o DataFrame resultante da função noiseCreation será atribuído à variável df
```

```python
import os 
files = os.listdir(path) # listar arquivos no diretório especificado por "path"
print(files)

# lista para armazenar todos os dados dos arquivos CSV
allData = [] # criando uma lista 
for file in files: # iterar sobre cada arquivo
    if file.endswith('.csv'): # verificar se o arquivo é um arquivo CSV
        print(file)
        data = pd.read_csv(path+'/'+file) # ler o arquivo CSV e adicionar os dados à lista 'allData'
        allData.append(data)
```

```python
allDataDf = pd.concat(allData) # concatenando todos os DataFrames presentes na lista "allData"
```
