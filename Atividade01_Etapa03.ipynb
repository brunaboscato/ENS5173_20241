# Atividade
#### Gerar números aleatórios que representem a atividade industrial, utilizando alguma indústria do estado para basear seus códigos.

Baseado na atividade industrial da BRF que produz em média 4 milhões de toneladas de alimentos por ano e no lucro que obteve no ultimo trimestre de 2023 de R$ 823 milhões. 

```python
import random
import pandas as pd
import matplotlib.pyplot as plt

# Definindo os valores médios
producao_anual = 4000000  # 4 milhões de toneladas
lucro_trimestral = 823000000  # R$ 823 milhões
```

```python
# Gerando números aleatórios para representar a produção mensal e lucro mensal
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
producao_mensal = [random.randint(300000, 500000) for _ in range(len(meses))]  # Geração de números aleatórios entre 300 mil e 500 mil toneladas
lucro_mensal = [random.randint(200000000, 300000000) for _ in range(len(meses))]  # Geração de números aleatórios entre R$ 200 milhões e R$ 300 milhões
```

```python
# Criando o DataFrame com os dados gerados
df = pd.DataFrame({
    'Produção (Toneladas)': producao_mensal,
    'Lucro (R$)': lucro_mensal 
}, index=meses)  # Definindo o índice como os meses
```

```python
# Adicionando total anual para produção e lucro
df.loc['Total Anual'] = [sum(producao_mensal), sum(lucro_mensal)]
print(df)
```

```python
df
```

```python
# Criando uma cópia do DataFrame excluindo a linha de total
df_plot = df.drop('Total Anual') 
df_plot_producao = df_plot['Produção (Toneladas)']
df_plot_lucro = df_plot['Lucro (R$)']
```

```python
# Plotando o gráfico
ax = df_plot_producao.plot(kind='bar', figsize=(10, 6))
ax.set_title('Produção')
ax.set_xlabel('Mês')
ax.set_ylabel('Quantidade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotando o gráfico
ax = df_plot_lucro.plot(kind='bar', figsize=(10, 6), color='green')
ax.set_title('Lucro')
ax.set_xlabel('Mês')
ax.set_ylabel('Quantidade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
