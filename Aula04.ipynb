```python
# Criando uma lista em python
list = [0,'eu amo polAtm',2.5] # 'eu amo polAtm' é uma str
print(list)
# Nesta parte fiz a adição entre duas strings, sendo que list[1] é o segundo indice do vetor list
print('variável:'+list[1])

# Adição do primeiro indice da variável list (list[0]) - lembrar que a indexação no python inicia em 0
print(list[0]+1)
```

```python
# Criando uma matriz numpy
import numpy as np

# Criando uma matriz de zeros utilizando np.zeros
a = np.zeros((5,5))
print(a)
print('Indexação numpy')
print(a[0,0])
```

```python
# Criando números aleatórios e mostrando a dimensão da matriz
b = np.random.rand(5,5) # uma matriz de 5 por 5
print(b)
print('Dimensão da matriz')
print(b.shape)
```

```python
# Criando matris de valores aleatórios entre uma faixa estabelecida
vmin = 1 # valor mínimo 
vmax = 100 # valor máximo

c = vmin+(vmax-vmin)*b
print(c)
```

```python
# Plotando a matriz
import matplotlib.pyplot as plt

plt.plot(c.transpose()) # transpose é a matriz transposta de c
```

```python
# Plotando apenas a primeira linha
x = np.array(range(0,np.size(c[0,:]))) # size é o tamanho da matriz e o array converte o range 
print(x)
plt.plot(x,c[0,:]) # : é todas as coluna
plt.title('Gráfico com Valores Inteiros no Eixo X')
plt.xticks(x) # definindo os valores do eixo x
```

```python
# Criando um gráfico de barras e exportanto

fig, ax = plt.subplots(3,1) # criou uma figura com 3 eixos
ax[0].bar(x,c[:,0]) # bar significa um gráfico de barra
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

# exportar a figura
import os
rootFolder = os.getcwd()
print(rootFolder)
figFolder = rootFolder.replace('\\','/')+'/figures/'
print(figFolder)
fig.savefig(figFolder+'Figure01.png')
```
